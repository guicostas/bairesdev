
### Notebook para Redução de Dimensionalidade em Imagens (Conversão para Tons de Cinza e Binarização)


```python
# Redução de Dimensionalidade em Imagens
# Conversão de Imagens Coloridas para Tons de Cinza e Binarização

# Instalando as bibliotecas necessárias (se já estiverem instaladas, o comando será ignorado)
!pip install opencv-python-headless numpy matplotlib

# Importando as bibliotecas
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Função para carregar e exibir imagens
def exibir_imagem(imagem, titulo="Imagem"):
    plt.imshow(imagem, cmap='gray' if len(imagem.shape) == 2 else None)
    plt.title(titulo)
    plt.axis('off')
    plt.show()

# 1. Carregar a imagem colorida
# Substitua 'caminho/para/sua/imagem.jpg' pelo caminho da sua imagem
caminho_imagem = 'dados/sample.png'
imagem_colorida = cv2.imread(caminho_imagem)

# Verificar se a imagem foi carregada corretamente
if imagem_colorida is None:
    print("Erro ao carregar a imagem. Verifique o caminho.")
else:
    # Converter a imagem de BGR (padrão do OpenCV) para RGB (para exibição correta com Matplotlib)
    imagem_colorida_rgb = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

    # Exibir a imagem colorida
    exibir_imagem(imagem_colorida_rgb, "Imagem Colorida")

    # 2. Converter a imagem para tons de cinza
    imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

    # Exibir a imagem em tons de cinza
    exibir_imagem(imagem_cinza, "Imagem em Tons de Cinza")

    # 3. Binarizar a imagem (preto e branco)
    # Definir um limiar (threshold) para a binarização
    limiar = 128  # Você pode ajustar esse valor conforme necessário
    _, imagem_binaria = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)

    # Exibir a imagem binarizada
    exibir_imagem(imagem_binaria, "Imagem Binarizada")

    # 4. Salvar as imagens processadas
    cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
    cv2.imwrite('imagem_binaria.jpg', imagem_binaria)

    print("Imagens salvas com sucesso: 'imagem_cinza.jpg' e 'imagem_binaria.jpg'")
```

### Explicação do Notebook:

-   **Bibliotecas**:
    -   `opencv-python-headless`: Para processamento de imagens.
    -   `numpy`: Para manipulação de arrays.
    -   `matplotlib`: Para exibição das imagens.
-   **Função `exibir_imagem`**:
    -   Exibe a imagem com um título e remove os eixos para melhor
        visualização.
-   **Conversão de cores**:
    -   O OpenCV carrega imagens no formato BGR, mas o `matplotlib` usa
        RGB. Por isso, convertemos a imagem para RGB antes de exibi-la.
-   **Binarização**:
    -   Usamos `cv2.threshold()` para criar uma imagem binária (preto e
        branco) com base em um limiar.
-   **Salvamento das imagens**:
    -   As imagens processadas são salvas como `imagem_cinza.jpg` e
        `imagem_binaria.jpg`.
